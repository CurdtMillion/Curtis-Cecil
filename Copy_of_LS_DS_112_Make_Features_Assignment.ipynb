```python
# Clean up the emp_title column
# First check out the top 20 employment titles
df['emp_title'].value_counts(dropna=False)[:20]
```

```python
# Examples of cases needed cleaned up
examples = ['owner', 'Supervisor', ' Project Manager', np.nan]
```

```python
# Function to clean up cases and increase uniformity
def clean_title(job_title):
  if isinstance(job_title, str):
    return job_title.title().strip()
  else:
    return "Unknown"
```

```python
# For loop
empty_list=[]
for title in examples:
  empty_list.append(clean_title(title))
```

```python
# List comrehensions combine function calls and for loops into one
# single line of code
[clean_title(title) for title in examples]
```

```python
# Apllying new function to column
df['emp_title']=df['emp_title'].apply(clean_title)
df.head()
```

```python
# Checking progress, looking at top 20 employment titles
df['emp_title'].value_counts(ascending=False)[:20]
```

```python
# Working with dates and times
# Using .dt accessor for dates with a for loop
mylist=[]
for col in df.columns:
  if col.endswith('_d'):
    mylist.append(col)
```

```python
# List comprehension
mycols = [col for col in df.columns if col.endswith('_d')]
```

```python
# String format of issue_d
df['issue_d'][:10]
```

```python
# Pandas will detect format of %m-%y and will translate it
df['issue_d'] = pd.to_datetime(df['issue_d'], infer_datetime_format=True)
df[mycols].head()
```

```python
# Now the month
df['month']=df['issue_d'].dt.month
df.head()
```
